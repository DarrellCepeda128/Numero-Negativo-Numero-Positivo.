```python
# Calculando Negativo o Positivo
# datos de entrada R , R
# mostrar numero
# si es verdadero mostrar Negativo
R = float(input("Ingrese un numero negativo: "))
R = float(input("Ingrese un numero positivo: "))
if R > 0:
    print("El numero es negativo")
    print(R)
elif R < 0:
    print("El numero es positivo")
    print(R)
else:
    print("El numero es cero")
    print(R)
print("Fin del programa")
print("Programa realizado por Cepeda Darrell NL = 10")
```
